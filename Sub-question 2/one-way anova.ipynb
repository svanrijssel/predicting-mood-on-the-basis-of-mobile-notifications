```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway, ttest_rel, wilcoxon
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
onehot_bored_f1 = [0.202, 0.1481, 0.1718, 0.1759, 0.1675, 0.0922, 0.1513, 0.2103, 0.2323, 0.1585]
binary_bored_f1 = [0.188, 0.1089, 0.1128, 0.0648, 0.1272, 0.1126, 0.1735, 0.1056, 0.154, 0.093]
r1_bored_f1 = [0.0779, 0.0968, 0.0818, 0.1023, 0.0764, 0.0989, 0.0875, 0.0971, 0.0829, 0.0883]
r2_bored_f1 = [0.0852, 0.0913, 0.0836, 0.0922, 0.0763, 0.093, 0.0838, 0.0925, 0.0862, 0.0908]
fh_bored_f1 = [0.0853, 0.1045, 0.0714, 0.087, 0.0783, 0.0925, 0.0883, 0.0773, 0.0908, 0.0893]

print(f_oneway(onehot_bored_f1, binary_bored_f1, r1_bored_f1, r2_bored_f1, fh_bored_f1))

print("Onehot_bored_f1 Mean & SD: ", round(np.mean(onehot_bored_f1), 4), round(np.std(onehot_bored_f1), 4))
print("binary_bored_f1 Mean & SD: ", round(np.mean(binary_bored_f1), 4), round(np.std(binary_bored_f1), 4))
print("r1_bored_f1 Mean & SD: ", round(np.mean(r1_bored_f1), 4), round(np.std(r1_bored_f1), 4))
print("r2_bored_f1 Mean & SD: ", round(np.mean(r2_bored_f1), 4), round(np.std(r2_bored_f1), 4))
print("fh_bored_f1 Mean & SD: ", round(np.mean(fh_bored_f1), 4), round(np.std(fh_bored_f1), 4))
```

```python
bored_f1_df = pd.DataFrame(data=[onehot_bored_f1, binary_bored_f1, r1_bored_f1, r2_bored_f1, fh_bored_f1]).transpose()
bored_f1_df.columns = ['One-hot', 'Binary', 'R1', 'R2', 'Frank&Hall']
bored_f1_df
ax = sns.boxplot(y="variable", x="value", data=pd.melt(bored_f1_df))
plt.xlabel("MA-F1 score")
plt.ylabel("Encoding scheme")
plt.title("Macro-averaged F1 scores for Bored")
plt.show(ax)

```
